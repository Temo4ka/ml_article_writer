```python
import torch
from unsloth import FastLanguageModel

model_name = "unsloth/Qwen3-4B"
max_seq_length = 2048

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=model_name,
    max_seq_length=max_seq_length,
    load_in_4bit=True,
    dtype=None,
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=32,
    lora_alpha=64,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    use_gradient_checkpointing="unsloth"
)
```

```python
from datasets import load_dataset

dataset = load_dataset("json", data_files=dataset_path, split="train")

def format_chat(example):
    text = tokenizer.apply_chat_template(
        example["messages"],
        tokenize=False,
        add_generation_prompt=False,
    )

    return {"text": text}

dataset = dataset.map(format_chat)
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    args=SFTConfig(
        output_dir="./qwen3-style-output",
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        num_train_epochs=3,
        learning_rate=2e-4,
        logging_steps=10,
        save_steps=50,
        optim="adamw_8bit",
        warmup_steps=5,
        lr_scheduler_type="linear",
        report_to="none",
        max_seq_length=max_seq_length,
        dataset_text_field="text",
    ),
)
```

```python
trainer.train()
```

```python
model.save_pretrained("qwen3-style-lora")
tokenizer.save_pretrained("qwen3-style-lora")

merged_model = model.merge_and_unload()
merged_model.save_pretrained("qwen3-style-merged")
tokenizer.save_pretrained("qwen3-style-merged")
```

```python
import gc
import torch

gc.collect()
torch.cuda.empty_cache()

from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    "qwen3-style-lora",
    max_seq_length=2048,
    load_in_4bit=True,
)
```

```python
def generate_article(theme):
    messages = [
        {"role": "system", "content": "Ты - опытный автор, который пишет статьи в стиле, максимально похожем на предоставленные примеры."},
        {"role": "user", "content": f"Напиши статью на тему: {theme} в твоём фирменном стиле. Не добавляй имя автора."},
    ]
    text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=1024, temperature=0.7, top_p=0.9, repetition_penalty=1.05)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print(generate_article("будущее нейросетей в 2026 году"))
```
